```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
Ecom = pd.read_excel("../Pandas/Ecommerce.xlsx")
```

```python
Ecom
```

```python
sns.kdeplot(Ecom["Price (Rs.)"])
```

```python
sns.boxplot(Ecom["Price (Rs.)"])
```

```python
sns.rugplot(Ecom["Price (Rs.)"])
```

```python
sns.violinplot(Ecom["Price (Rs.)"])
```

```python
figure=plt.figure(figsize=(12, 6))
sns.countplot(x="Category", data=Ecom)
plt.title("Count of Category")
plt.show()
```

```python
sns.stripplot(x=Ecom["Price (Rs.)"])
plt.show()
```

```python
sns.displot(Ecom["Price (Rs.)"])
plt.show()
```

# Bivariate Charts

```python
sns.regplot(data=Ecom,x="Price (Rs.)", y="Discount (%)", scatter_kws={"color":"orange"},line_kws={"color":"blue"})
plt.show()
```

```python
sns.jointplot(data=Ecom, x="Price (Rs.)", y="Discount (%)",kind="kde")
plt.show()
```

```python
sns.jointplot(data=Ecom, x="Price (Rs.)", y="Discount (%)",kind="hex")
plt.show()
```

# MultiVariate Chart


```python
figure=plt.figure(figsize=(12,6))
sns.catplot(x="Category", y="Price (Rs.)",data=Ecom, kind="bar", hue="Payment_Method") # use third attribute hue
plt.show()
```

```python
figure=plt.figure(figsize=(12,6))
sns.relplot(x="Purchase_Date",y="Price (Rs.)",data=Ecom,hue="Category",style="Payment_Method")
```

```python
g=sns.FacetGrid(Ecom,col="Payment_Method",height=5)
g.map(sns.countplot,"Category")
plt.show()
```

```python
sns.pairplot(Ecom,hue="Category")
```

```python
sns.heatmap(Ecom[["Price (Rs.)","Discount (%)","Final_Price(Rs.)"]].corr())
```

```python
sns.clustermap(Ecom[["Price (Rs.)","Discount (%)","Final_Price(Rs.)"]].corr())
```
