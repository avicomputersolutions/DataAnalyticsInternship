```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = sns.load_dataset("tips")
```

```python
dataset
```

```python
#count Chart
sns.countplot(x="sex",data=dataset)
plt.title("No of Male and Female")
plt.xlabel("Gender")
plt.show()
```

```python
sns.countplot(x="smoker",data=dataset,palette="magma")
plt.title("No of Smoker or Non Smoker")
plt.xlabel("Smoker")
plt.show()
```

```python
sns.countplot(x="day",data=dataset,palette="rocket")
plt.title("Day")
plt.xlabel("Days")
plt.show()
```

```python
sns.countplot(x="time",data=dataset,palette="flare")
plt.title("People visited by time")
plt.xlabel("Time")
plt.show()
```

```python
dataset
```

```python
sns.swarmplot(x="day",y="tip",data=dataset,hue="sex")
plt.title("Day vs Tips")
plt.xlabel("Day")
plt.ylabel("Tips")
plt.show()
```

```python
sns.pointplot(x="day",y="tip",data=dataset,hue="sex")
plt.title("Day vs Tips")
plt.xlabel("Day")
plt.ylabel("Tips")
plt.show()
```

```python
sns.boxplot(x="day",y="tip",data=dataset,hue="sex")
plt.title("Day vs Tips")
plt.xlabel("Day")
plt.ylabel("Tips")
plt.show()
```

```python
sns.violinplot(x="day",y="tip",data=dataset,hue="sex")
plt.title("Day vs Tips")
plt.xlabel("Day")
plt.ylabel("Tips")
plt.show()
```

```python
sns.catplot(x="day",y="tip",data=dataset,kind="strip", hue="sex")
plt.title("Day vs Tips")
plt.xlabel("Day")
plt.ylabel("Tips")
plt.show()
```
